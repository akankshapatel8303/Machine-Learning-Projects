```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

### Let's apply `OLS`

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
pred = lr.predict(X)
pred
```

```python
plt.scatter(X,y)
plt.plot(X, pred, color="red")
```

### Lets apply Gradient Descent assuming slope is constant m = 78.35 
### and lets assume the starting value for intercept b = 0

```python
y_pred = ((78.35 * X) + 100).reshape(4)
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X, pred, color="red", label="OLS")
plt.plot(X, y_pred, color="green", label="b = 0")
plt.legend()
plt.show()
```

```python
m = 78.35
b = 100
# Iteration 1
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
# Let,s take learning rate = 0.1
lr = 0.1
step_size = loss_slope*lr
step_size
```

```python
# Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)
y_pred1
```

```python
plt.scatter(X,y)
plt.plot(X, pred, color="red", label="OLS")
plt.plot(X, y_pred1, color="#00a65a", label="b = {}".format(b))
plt.plot(X, y_pred, color="#A3E4D7", label="b = 0")
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35*X + b)).reshape(4)
y_pred2
```

```python
plt.scatter(X,y)
plt.plot(X, pred, color="red", label="OLS")
plt.plot(X, y_pred2, color="#00a65a", label="b = {}".format(b))
plt.plot(X, y_pred1, color="#A3E4D7", label="b = {}".format(b))
plt.plot(X, y_pred, color="#A3E4D7", label="b = 0")
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() -b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = (m*X + b).reshape(4)
y_pred3
```

```python
plt.scatter(X,y)
plt.plot(X, pred, color="red", label="OLS")
plt.plot(X, y_pred3, color="#00a65a", label="b = {}".format(b))
plt.plot(X, y_pred2, color="#A3E4D7", label="b = {}".format(b))
plt.plot(X, y_pred1, color="#A3E4D7", label="b = {}".format(b))
plt.plot(X, y_pred, color="#A3E4D7", label="b = 0")
plt.legend()
plt.show()
```

```python
# Iteration 4
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred4 = (m*X + b).reshape(4)
y_pred4
```

```python
plt.scatter(X,y)
plt.plot(X, pred, color="red", label="OLS")
plt.plot(X, y_pred4, color="#00a65a", label="b = {}".format(b))
plt.plot(X, y_pred3, color="#A3E4D7", label="b = {}".format(b))
plt.plot(X, y_pred2, color="#A3E4D7", label="b = {}".format(b))
plt.plot(X, y_pred1, color="#A3E4D7", label="b = {}".format(b))
plt.plot(X, y_pred, color="#A3E4D7", label="b = 0")
plt.legend()
plt.show()
```

# SHORT METHOD

```python
b = -100
lr = 0.1
m = 78.35
epochs = 15

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    step_size = loss_slope*lr
    b = b - step_size
    y_pred = m*X + b

    plt.plot(X, y_pred)

plt.scatter(X, y)
```
